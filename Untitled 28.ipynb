```python
# Funktion zur Entfernung von Ausreißern
def remove_outliers(df, column, lower_bound=None, upper_bound=None):
    if lower_bound is None or upper_bound is None:
        q1 = df[column].quantile(0.25)
        q3 = df[column].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"Ausreißer in {column}:")
    print(outliers)
    print()

    # Ersetzen der Ausreißer mit dem Durchschnitt
    mean = df[column].mean()
    df[column] = df[column].apply(lambda x: mean if x < lower_bound or x > upper_bound else x)

    return df

# Daten laden (Beispiel)
Kundendaten_03 = pd.DataFrame({'Einkommen': np.random.uniform(10000, 200000, 100)})
Besucherdaten_03 = pd.DataFrame({'Einkommen': np.random.uniform(10000, 200000, 100)})

# Bestimmen der unteren und oberen Grenzen für Ausreißer
lower_bound = Kundendaten_03['Einkommen'].min()
upper_bound = 150000  # Beispiel-Obergrenze

# Identifizieren und Bereinigen von Ausreißern
print("Bereinigung für Kundendaten_03 (Einkommen):")
Kundendaten_04 = remove_outliers(Kundendaten_03.copy(), 'Einkommen', lower_bound, upper_bound)

print("Bereinigung für Besucherdaten_03 (Einkommen):")
Besucherdaten_04 = remove_outliers(Besucherdaten_03.copy(), 'Einkommen', lower_bound, upper_bound)

# Histogramme erstellen
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Histogramm für Kundendaten_04
axs[0].hist(Kundendaten_04['Einkommen'], bins=np.arange(Kundendaten_04['Einkommen'].min(), Kundendaten_04['Einkommen'].max() + 10000, 10000), edgecolor='black')
axs[0].set_xticks(np.arange(Kundendaten_04['Einkommen'].min(), Kundendaten_04['Einkommen'].max() + 10000, 10000))
axs[0].set_xlabel('Einkommen')
axs[0].set_ylabel('Frequency')
axs[0].set_title('Einkommen der Kunden')

# Anzahl über den Säulen platzieren
counts_kunde, bins_kunde = np.histogram(Kundendaten_04['Einkommen'], bins=np.arange(Kundendaten_04['Einkommen'].min(), Kundendaten_04['Einkommen'].max() + 10000, 10000))
for count, bin in zip(counts_kunde, bins_kunde[:-1]):
    axs[0].text(bin + 5000, count + 0.05, str(int(count)), ha='center', va='bottom')

# Histogramm für Besucherdaten_04
axs[1].hist(Besucherdaten_04['Einkommen'], bins=np.arange(Besucherdaten_04['Einkommen'].min(), Besucherdaten_04['Einkommen'].max() + 10000, 10000), edgecolor='black')
axs[1].set_xticks(np.arange(Besucherdaten_04['Einkommen'].min(), Besucherdaten_04['Einkommen'].max() + 10000, 10000))
axs[1].set_xlabel('Einkommen')
axs[1].set_ylabel('Frequency')
axs[1].set_title('Einkommen der Besucher')

# Anzahl über den Säulen platzieren
counts_besucher, bins_besucher = np.histogram(Besucherdaten_04['Einkommen'], bins=np.arange(Besucherdaten_04['Einkommen'].min(), Besucherdaten_04['Einkommen'].max() + 10000, 10000))
for count, bin in zip(counts_besucher, bins_besucher[:-1]):
    axs[1].text(bin + 5000, count + 0.05, str(int(count)), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
